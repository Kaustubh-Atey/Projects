```
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from os import listdir
```

```
filenames = []
```

```
for file in listdir('/content/try'):
  filenames.append(file)
```

```
filenames
```

```
Image_ID = []
Content = []	
Address = []
Path = []
Type = []
```

```
for filename in filenames:
  response = open(filename)
  html_doc = response.read()

  soup = BeautifulSoup(html_doc, 'html.parser')

  for row in soup.find_all('tr'):
  
    for i in row.find_all('img'):
      Address.append(i)
      Content.append(row.contents[5].contents[0].contents[0]) 
      Type.append(row.contents[1].contents[0].contents[0].contents[0])

    image_id = [] 

    for i in range(len(row.find_all('img'))):
      index = -2*i - 2  
      image_id.append(row.contents[index].contents[2].contents[1].contents[1].contents[1].contents[1].contents[0])
      image_id = image_id[::-1]  

    Image_ID = Image_ID + image_id  
```

```
for i in Address:
    Path.append(i.attrs.get('src',''))
```

```
print(len(Path))
print(len(Content))
print(len(Image_ID))
print(len(Type))
```

```
reg = pd.DataFrame(columns = ['Type','Image_ID','Content','Path'])
reg['Type'] = Type
reg['Image_ID'] = Image_ID
reg['Content'] = Content
reg['Path'] = Path

reg.head()
```

```
results = reg.to_csv('results.csv')
```
