```
import numpy as np
import pandas as pd
import xml.etree.ElementTree as ET
```

```
filenames = []
```

```
Background = []
Image_ID = []
Content = []
```

```
from os import listdir
for file in listdir('/content/try'):
  filenames.append(file)

print(filenames)  
```

```
for filename in filenames:

    with open(filename, 'r', encoding="utf-8") as content:
      
      if filename.endswith('.xml'):
        print(filename, 'is .xml')

        tree = ET.parse(content)
        root = tree.getroot()

        image_id = root[0].text
        Image_ID.append(image_id)
        background = root[-3].text
        Background.append(background)
        category = root[-2].text
        Content.append(category)

      else:
        print(filename, 'is not .xml')  
```

```
print('Image ID:', Image_ID)
print('Content:', Content)
print('Background:', Background)
```

```
df = pd.DataFrame( columns = ['Image_ID','Content','Background'])
df['Image_ID'] = Image_ID
df['Content'] = Content
df['Background'] = Background
df.head()
```

```
a = df.to_csv('results.csv')
```
