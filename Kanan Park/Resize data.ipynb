## **Train**

```
Entire_path = '/content/drive/My Drive/Seperated files/Train/Entire'
Flower_path = '/content/drive/My Drive/Seperated files/Train/Flower'
Fruit_path = '/content/drive/My Drive/Seperated files/Train/Fruit'
Leaf_path = '/content/drive/My Drive/Seperated files/Train/Leaf'
Stem_path = '/content/drive/My Drive/Seperated files/Train/Stem'
Leaf_sheet_path = '/content/drive/My Drive/Seperated files/Train/Stembg'
```

```
Entire = os.listdir(Entire_path)
Flower = os.listdir(Flower_path)
Fruit = os.listdir(Fruit_path)
Leaf = os.listdir(Leaf_path)
Stem = os.listdir(Stem_path)
Leaf_sheet = os.listdir(Leaf_sheet_path)
```

```
# Resize the images
for file in Entire:
    im = Image.open(Entire_path + '/' + file)
    img = im.resize((100, 100))
    img.save('/content/drive/My Drive/KananPark/Dataset/Train/Entire' +'/' + file,'JPEG')
```

```
for file in Leaf:
    im = Image.open(Leaf_path + '/' + file)
    img = im.resize((100, 100))
    img.save('/content/drive/My Drive/KananPark/Dataset/Train/Leaf' +'/' + file,'JPEG')
```

```
for file in Stem:
    im = Image.open(Stem_path + '/' + file)
    img = im.resize((100, 100))
    img.save('/content/drive/My Drive/KananPark/Dataset/Train/Stem' +'/' + file,'JPEG')
```
